1. Nombres y créditos de clientes de los compradores de Baumbach Inc (76) ordenado por crédito descendente y nombre ascendente

```python
%%sql 
SELECT b.nombre,b.credito FROM compra as A
left join cliente as B on A.cliente_id=b.id 
where tienda_id=76
order by credito desc,nombre asc
```

2. Agregar en el resultado anterior el crédito en dólares a un tipo de cambio de 7.8.

```python
%%sql
SELECT b.nombre,b.credito,round(b.credito/7.8,2) as credito_dolares FROM compra as A
left join cliente as B on A.cliente_id=b.id 
where tienda_id=76
order by credito desc,nombre asc
```

3. Parejas de clientes con el mismo crédito que no sean ellos mismos y que no haya parejas duplicadas

```python
%%sql
select distinct c1.nombre , c2.nombre, c1.credito, c2.credito from cliente c1 left join cliente c2 on c1.id <>c2.id
where c1.credito=c2.credito ;
```

4. Clientes que han comprado en Baumbach Inc (76)  y no en Lang-Conn (5)

```python
%%sql 
SELECT b.nombre,b.credito FROM compra as A
left join cliente as B on A.cliente_id=b.id 
where tienda_id=76 and tienda_id<>5
```

5. Todas las tiendas que tengan otras tiendas en la misma dirección

```python
%%sql 
select distinct t1.nombre ,t1.direccion from tienda t1 left join tienda t2 on t1.id <>t2.id
where t1.direccion=t2.direccion ;
```

9. Diferencia entre el promedio de crédito de los clientes de Baumbach Inc (76) y de tienda con id (6)

```python
%%sql
select promedio_baum, promedio_lang, abs(promedio_baum-promedio_lang) diferencia from (
(SELECT avg(b.credito) promedio_baum FROM compra as A
left join cliente as B on A.cliente_id=b.id 
where tienda_id=76 )
inner join (
SELECT avg(b.credito) promedio_lang FROM compra as A
left join cliente as B on A.cliente_id=b.id 
where tienda_id=6) on 1=1 )
```

10. Número de clientes de cada tienda, de las que tienen compras, ordenado por nombre de tienda.

```python
%%sql
select t.nombre, count(c.cliente_id) num_clientes from compra as c left join tienda t on c.tienda_id=t.id
group by t.nombre
order by nombre
```
